```python
# Drop columns Area_Type, HOUSE_TYPE, Purpose (As all values are same except 2 and that won't affect prediction)
# DataFrame.drop() if inplace=false,return a copy.Otherwise, do operation inplace ansd return None.
house_data.drop(['Area_Type', 'HOUSE_TYPE', 'Purpose', 'Owner_name'], axis=1, inplace=True)
house_data.head()
```

```python
# Trimming /sq.ft from right of Price_per_sq.ft and sq.ft from right of Total_Sq.ft
# rstrip : string method สำหรับตัวอักขระต่อท้ายของ string
house_data["Price_per_sq.ft"] = house_data.apply(lambda x: x["Price_per_sq.ft"].rstrip("/sq.ft"), axis = 1)
house_data["Total_Sq.ft"] = house_data.apply(lambda x: x["Total_Sq.ft"].rstrip(" sq.ft"), axis = 1)

# Trimming ₹ from left of Flat_Price, EMI_Starts, Price_per_sq.ft
house_data["Flat_Price"] = house_data.apply(lambda x: x["Flat_Price"].lstrip("₹"), axis = 1)
house_data["EMI_Starts"] = house_data.apply(lambda x: x["EMI_Starts"].lstrip("₹"), axis = 1)
house_data["Price_per_sq.ft"] = house_data.apply(lambda x: x["Price_per_sq.ft"].lstrip("₹"), axis = 1)

house_data.head(10)
```

```python
# Checking unique values after space in below columns and comma(,) for Location
# DataFrame.str.split(' ',1,expand=True); 
house_data['Flat_Price'].str.split(' ',1, expand=True)[1].unique()
house_data['EMI_Starts'].str.split(' ', 1, expand=True)[1].unique()
house_data['Price_per_sq.ft'].str.split(' ', 1, expand=True)[1].unique()
house_data['Location'].str.split(',', 1, expand=True)[1].unique()
house_data['BHK'].str.split(' ', 1, expand=True)[1].unique()
```

```python
# Splitting Location into Region and City and then dropping City column as it contains only Kolkata value and also Location column

house_data[['Region', 'City']] = house_data['Location'].str.split(', ', 1, expand=True)
house_data.drop(['City', 'Location'], axis=1, inplace=True)

# Splitting BHK into Rooms and BHK_type and then dropping BHK column
house_data[['Rooms', 'BHK_type']] = house_data['BHK'].str.split(' ', 1, expand=True)
house_data.drop(['BHK'], axis=1, inplace=True)

house_data.head()
```

```python
# Getting the unexpected value rows for Flat_Price
house_data[(house_data['Flat_Price'].str.split(' ', 1, expand=True)[1] == 'L - 1.03 Cr')]
```

```python
# Dropping rows 148 and 168 as almost many values are Null
house_data = house_data.drop([148, 168], axis=0)
```

```python
# Method to convert Flat_Price, EMI_Starts, Price_per_sq.ft to integers
def convert_price_to_integer(row_value):
    splitted_row_value = row_value.split(" ")
    amount = float(splitted_row_value[0])
    if len(splitted_row_value) == 2:
        unit = splitted_row_value[1]
        if unit == 'K':
            return 1000*float(amount)
        elif unit == 'L' or unit == 'Lacs':
            return 100000*float(amount)
        elif unit == 'Cr':
            return 10000000*float(amount)
    else:
        return int(amount)
    
house_data["Price_per_sq.ft"] = house_data.apply(lambda x: convert_price_to_integer(x["Price_per_sq.ft"]), axis = 1)
house_data["Flat_Price"] = house_data.apply(lambda x: convert_price_to_integer(x["Flat_Price"]), axis = 1)
house_data["EMI_Starts"] = house_data.apply(lambda x: convert_price_to_integer(x["EMI_Starts"]), axis = 1)

house_data["Total_Sq.ft"] = house_data.apply(lambda x: (float(x["Total_Sq.ft"])), axis = 1)
house_data["Rooms"] = house_data.apply(lambda x: (float(x["Rooms"])), axis = 1)

house_data.head()
```

```python
# Converting Owner_type, Region and BHK_type to integer values
owner_types = house_data["Owner_type"].unique().tolist()
house_data["Owner_type"] = [owner_types.index(value) for value in house_data["Owner_type"].to_list()]

regions = house_data["Region"].unique().tolist()
house_data["Region"] = [regions.index(value) for value in house_data["Region"].to_list()]

BHK_types = house_data["BHK_type"].unique().tolist()
house_data["BHK_type"] = [BHK_types.index(value) for value in house_data["BHK_type"].to_list()]

house_data.head()
```

```python
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
```

```python
# y -> dependent variable
X = house_data.drop("EMI_Starts", axis = 1)
y = house_data["EMI_Starts"]
X.head()
y.head()
```

```python
# Split dataset into training set and test set
# test_sizefloat or int, default=None
# If float, should be between 0.0 and 1.0 and represent the proportion of the dataset to include in the test split. 
# If int, represents the absolute number of test samples. 
# If None, the value is set to the complement of the train size. If train_size is also None, it will be set to 0.25.

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0, shuffle = True) # 80% training and 20% test
```

```python
# Method for checking regression model performances
def performance_check(prediction):
    print("Mean Square Error: {}\nMean Absolute Error: {}\nR2 Score: {}\n"
          .format(metrics.mean_squared_error(y_test, prediction), 
                  metrics.mean_absolute_error(y_test, prediction),
                  metrics.r2_score(y_test, prediction)))
```

## Linear Regression

```python
# importing module
from sklearn.linear_model import LinearRegression

# creating an object
linearRegression = LinearRegression()

# fitting the training data
linearRegression.fit(x_train,y_train)

# Predict EMI on test values
y_prediction =  linearRegression.predict(x_test)

performance_check(y_prediction)
```

## K Nearest Neighbors

```python
#import required packages
from sklearn.preprocessing import MinMaxScaler
from sklearn import neighbors

scaler = MinMaxScaler(feature_range=(0, 1))

x_train_scaled = scaler.fit_transform(x_train)
x_train_scaled = pd.DataFrame(x_train_scaled)

x_test_scaled = scaler.fit_transform(x_test)
x_test_scaled = pd.DataFrame(x_test_scaled)

for K in range(15, 25):
    K = K+1
    model = neighbors.KNeighborsRegressor(n_neighbors = K)

    model.fit(x_train_scaled, y_train)  #fit the model
    y_prediction=model.predict(x_test_scaled) #make prediction on test set
    performance_check(y_prediction)
```
